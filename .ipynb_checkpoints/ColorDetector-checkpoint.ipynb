```python
import tensorflow as tf
import os
```

```python
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)
```

```python
import cv2
import imghdr
from matplotlib import pyplot as plt
```

```python
data_dir = "data"
```

```python
img_ext = ["jpg", "jpeg", "png", "bmp"]
```

```python
for color in os.listdir(data_dir):
    for img in os.listdir(os.path.join(data_dir, color)):
        img_path = os.path.join(data_dir, color, img)
        try:
            im = cv2.imread(img_path)
            end = imghdr.what(img_path)
            if end not in img_ext:
                print("Extension error with {}".format(img_path))
                os.remove(img_path)
        except Exception as e:
            print("Issue with image {}".format(img_path))
```

```python
import numpy as np
```

```python
data = tf.keras.utils.image_dataset_from_directory("data", label_mode='categorical')
```

```python
dataiter = data.as_numpy_iterator()
```

```python
batch1 = dataiter.next()
```

```python
batch1[1]
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch1[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch1[1][idx])
```

```python
# [blue, green, orange, purple, red, yellow]
```

```python
data = data.map(lambda x,y: (x/255,y))
```

```python
scalitr = data.as_numpy_iterator()
```

```python
batch1 = scalitr.next()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch1[0][:4]):
    ax[idx].imshow(img)
    ax[idx].title.set_text(batch1[1][idx])
```

```python
len(data)
```

```python
train_sz = 20
valid_sz = 6
test_sz = 3
```

```python
train_data = data.take(train_sz)
val_data = data.skip(train_sz).take(valid_sz)
test_data = data.skip(train_sz + valid_sz).take(test_sz)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation="relu", input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation="relu"))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation="relu"))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation="relu"))
model.add(Dense(6, activation="sigmoid"))
```

```python
model.compile('adam', loss = tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
ldir = "logs"
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = ldir)
```

```python
hist = model.fit(train_data, epochs = 20, validation_data = val_data, callbacks = [tensorboard_callback])
```

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
Pre = Precision()
Rec = Recall()
Acc = BinaryAccuracy()
```

```python
for batch in test_data.as_numpy_iterator():
    x,y = batch
    y_hat = model.predict(x)
    Pre.update_state(y, y_hat)
    Rec.update_state(y, y_hat)
    Acc.update_state(y, y_hat)
```

```python
print(Pre.result(), Rec.result(), Acc.result())
```
